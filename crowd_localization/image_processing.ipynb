### 졸업논문


```python
import zipfile
import os
import shutil
from PIL import Image
```

```python
folder_path = "../ProcessedData/AIhub/images/Training/"

file_list = os.listdir(folder_path)
zip_file_list = []

for file_name in file_list:
    if file_name.endswith(".zip"):
        print(file_name)
        zip_file_list.append(file_name)
```

```python
DATA_NAME = "TS_2.시나리오_143.Outdoor_일도체육공원143(593)"
```

```python
zip_file_path = "../ProcessedData/AIhub/images/Training/" + DATA_NAME + ".zip"
extraction_dir = "../ProcessedData/AIhub/images/Training/" + DATA_NAME

os.makedirs(extraction_dir, exist_ok=True)

with zipfile.ZipFile(zip_file_path, "r") as zip_ref:
    zip_ref.extractall(extraction_dir)

print(f"Contents of {zip_file_path} extracted to {extraction_dir}")
```

```python
source_folder = "../ProcessedData/AIhub/images/Training/" + DATA_NAME
destination_folder = os.path.join(source_folder, "original_images")

os.makedirs(destination_folder, exist_ok=True)

file_list = os.listdir(source_folder)

for file_name in file_list:
    source_file = os.path.join(source_folder, file_name)
    destination_file = os.path.join(destination_folder, file_name)

    if os.path.isdir(source_file):
        continue

    shutil.move(source_file, destination_file)

print("All files moved to the 'original_images' folder.")
```

```python
folder_path = "../ProcessedData/AIhub/images/Training/" + DATA_NAME + "/original_images"
num_start = 1

file_list = os.listdir(folder_path)
file_list.sort()

for idx, file_name in enumerate(file_list, start=num_start):
    old_file_path = os.path.join(folder_path, file_name)
    new_file_name = f"{idx}.jpg"
    new_file_path = os.path.join(folder_path, new_file_name)

    os.rename(old_file_path, new_file_path)

print("All files in the 'original_images' folder have been renamed.")
```

```python
output_file = "../ProcessedData/AIhub/images/Training/" + DATA_NAME + "/list.txt"
with open(output_file, "w") as f:
    for i in range(340, 376):
        f.write(str(i) + "\n")
```

```python
source_folder = (
    "../ProcessedData/AIhub/images/Training/" + DATA_NAME + "/original_images"
)
destination_folder = "../ProcessedData/AIhub/images/Training/" + DATA_NAME + "/images"

os.makedirs(destination_folder, exist_ok=True)

new_size = (1280, 720)

file_list = os.listdir(source_folder)

for file_name in file_list:
    source_file = os.path.join(source_folder, file_name)
    destination_file = os.path.join(destination_folder, file_name)

    if os.path.isfile(source_file) and file_name.lower().endswith(
        (".png", ".jpg", ".jpeg", ".bmp", ".tiff")
    ):
        with Image.open(source_file) as img:
            resized_img = img.resize(new_size)
            resized_img.save(destination_file)

print("All images resized and saved to the 'images' folder.")
```

### 전국과학전람회


```python
import cv2
import os

TEST_DATA_NAME = ""
TEST_DATA_ROOT = f"../ProcessedData/{TEST_DATA_NAME}"

video_path = f"{TEST_DATA_ROOT}/video.mp4"
output_path = f"{TEST_DATA_ROOT}/images"

os.makedirs(output_path, exist_ok=True)

video_capture = cv2.VideoCapture(video_path)
frame_count = 0

desired_width = 1280
desired_height = 720

while True:
    ret, frame = video_capture.read()

    if not ret:
        break

    resized_frame = cv2.resize(frame, (desired_width, desired_height))

    image_path = f"{output_path}/{frame_count}.jpg"
    cv2.imwrite(image_path, resized_frame)

    with open(f"{TEST_DATA_ROOT}/list.txt", "a") as f:
        f.write(f"{frame_count}\n")

    frame_count += 1

video_capture.release()
cv2.destroyAllWindows()
```

```python
from moviepy.editor import VideoFileClip, AudioFileClip

TEST_DATA_NAME = ""
TEST_DATA_ROOT = f"../ProcessedData/{TEST_DATA_NAME}"

input_mov_file = f"{TEST_DATA_ROOT}/video.mov"
output_mp4_file = f"{TEST_DATA_ROOT}/video.mp4"


def convert_mov_to_mp4(input_path, output_path):
    try:
        video_clip = VideoFileClip(input_path)

        print(f"Video duration: {video_clip.duration}")
        print(f"Video fps: {video_clip.fps}")
        print(f"Video size: {video_clip.size}")

        audio_clip = video_clip.audio if video_clip.audio else None

        video_clip.write_videofile(output_path, codec="libx264", audio=False)

        if audio_clip:
            video_with_audio = video_clip.set_audio(audio_clip)
            video_with_audio.write_videofile(output_path, codec="libx264")

    except Exception as e:
        print(f"An error occurred: {e}")


convert_mov_to_mp4(input_mov_file, output_mp4_file)
print(f"Conversion complete: {output_mp4_file}")
```
